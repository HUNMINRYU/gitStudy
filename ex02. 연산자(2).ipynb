```python
# 숫자와 문자열 더하기

num1 = 10
str2 = "7"

# 형변환을 통해서 정수를 문자열로, 문자열을 정수로 변경해줄 수 있음!
# int() : int형으로 강제형변환
print(num1 + int(str2))

# type() : 타입 확인
print(type(str2))

# str() : str형으로 강제형변환
print(str(num1) + str2)
```

```python
# 문자열 곱셈(횟수만큼 이어붙인다!)
'안녕하세요' * 8
```

```python
num1, num2 = 23, 3
print(f"더하기 결과 : {num1 + num2}")
print(f"빼기 결과 : {num1 - num2}")
print(f"곱하기 결과 : {num1 * num2}")
print(f"나누기 결과 : {num1 / num2}")
print(f"몫 결과 : {num1 // num2}")
print(f"나머지 결과 : {num1 % num2}")
print(f"제곱 결과 : {num1 ** num2}")
```

```python
# input() 함수
# 키보드로 입력할 수 있는 함수
# 자바의 Scanner와 같은 기능

# ※ input() 함수 주의사항!
# 1. input 함수의 반환값은 문자열(String)형이다!
# 2. 연산을 수행해야 할 경우, 형변환이 필요하다!
# 3. 입력을 수행해야지만 다음 코드로 진행 가능!

num1 = int(input("첫번째 정수를 입력해주세요 >> "))
num2 = int(input("두번째 정수를 입력해주세요 >> "))

print(f"더하기 결과 : {num1 + num2}")
print(f"빼기 결과 : {num1 - num2}")
print(f"곱하기 결과 : {num1 * num2}")
print(f"나누기 결과 : {num1 / num2}")
print(f"몫 결과 : {num1 // num2}")
print(f"나머지 결과 : {num1 % num2}")
print(f"제곱 결과 : {num1 ** num2}")
```

```python
# 삼항연산자
# a if 조건식 else b
# 조건식이 True이면 a 실행(왼쪽)
# 조건식이 False이면 b 실행(오른쪽)

score = 55
'합격' if score >= 60 else '불합격'
```

```python
if score >= 60 :
    '합격'
else :
    '불합격'
```
