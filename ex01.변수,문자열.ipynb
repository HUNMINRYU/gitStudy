#### 문자열 인덱싱, 슬라이싱
- 시퀀스 자료구조(순서가 있는 자료구조)에 인덱스가 존재!
- 인덱스(index): 데이터가 생성될 때 자동으로 부여되는 순서 번호
- 인덱싱(indexing): 무엇인가(요소 하나)를 가리킨다는 의미
- 슬라이싱(slicing): 무엇(요소 여러개)를 잘라낸다는 의미

```python
#문자열 인덱싱
s = "My name is HM"

print(s[0],s[1],s[2],s[3],s[4],s[5],s[6],s[7],s[8],s[9],s[10],s[11],s[12])
```

```python
print(s[-1])
```

```python
#문자열 슬라이싱(⚠ 시작값은 포함, 끝값은 미포함⚠)
print(s[0:2]) #s[0]~s[1]까지 자름
print(s[3:7]) #s[3]~s[6]까지 자름
print(s[-5:-3]) #s[-3]~s[-5]까지 자름
```

```python
# 슬라이싱 문법
#변수명[시작할값(포함):끝값(미포함)]
#시작값이나 끝값을 비워두게 되면 → "처음"부터 "끝"까지
# 'My name'
print(s[:7]) # 처음부터 ~

#'name is HM'
print(s[3:]) # ~ 끝까지
```

```python
# 문자열 슬라이싱 실습
dy_info = "2023년 11월 29일의 날씨는 맑음입니다."
year = dy_info[:5]
day = dy_info[6:13]
weather = dy_info[19:21] #[-6:-4]

#문자열 이을 때 +(매개변수1개) 또는 ,(매개변수 여러개) 활용
print("년도:", year)
print("월일:", day)
print("날씨:", weather)
```

##### 문자열 포매팅(Formatting)
- 문자열 안의 특정한 값을 바꿔야(삽입)할 경우가 있을 때 사용

```python
# ctrl + shift + '-'                     : 셀 분할
# shift 누른 상태에서 셀 선택 후 shift + m : 셀 병합
```

```python
# %기호 포매팅
month = 11
day = 29

s = "오늘은 %d월 %d일입니다." %(month, day)
s
```

#### 문자열 포맷 코드


```python
# format함수 포매팅
month = 11
day = 29

s = "오늘은 {}월 {}일입니다." .format(month, day)
s
```

```python
# f문자열 포매팅
month = 11
day = 29

s = f"오늘은 {month}월 {day}일입니다." 
s
```
